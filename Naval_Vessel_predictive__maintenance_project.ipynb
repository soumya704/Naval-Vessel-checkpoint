```python
x=df.iloc[:,:-3]
y=df.iloc[:,-2:-1]
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
x_scaled = scaler_x.fit_transform(x)
y_scaled = scaler_y.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf1=DecisionTreeRegressor()
clf1.fit(x_train,y_train)
```

```python
y_pred1=clf1.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_absolute_percentage_error
```

```python
print('MSE:', mean_squared_error(y_test, y_pred1))
print('R2 Score:', r2_score(y_test, y_pred1))
print('MAE:', mean_absolute_error(y_test, y_pred1))
print('MAPE:', mean_absolute_percentage_error(y_test, y_pred1))
```

```python
from sklearn.ensemble import RandomForestRegressor
clf2=RandomForestRegressor()
clf2.fit(x_train,y_train)
```

```python
y_pred2=clf2.predict(x_test)
```

```python
print('MSE:', mean_squared_error(y_test, y_pred2))
print('R2 Score:', r2_score(y_test, y_pred2))
print('MAE:', mean_absolute_error(y_test, y_pred2))
print('MAPE:', mean_absolute_percentage_error(y_test, y_pred2))
```

```python
from sklearn.linear_model import LinearRegression
clf3=LinearRegression()
clf3.fit(x_train,y_train)
```

```python
y_pred3=clf3.predict(x_test)
```

```python
print('MSE:', mean_squared_error(y_test, y_pred3))
print('R2 Score:', r2_score(y_test, y_pred3))
print('MAE:', mean_absolute_error(y_test, y_pred3))
print('MAPE:', mean_absolute_percentage_error(y_test, y_pred3))
```

```python
models=['Decision Tree','Random Forest','Linear Regression']
mse_values=[mean_squared_error(y_test, y_pred1),mean_squared_error(y_test, y_pred2),mean_squared_error(y_test, y_pred3)]
```

```python
plt.figure()
plt.bar(models, mse_values, color=['Blue','Green','Red'])
plt.xlabel('Models')
plt.ylabel('MSE')
plt.title('MSE Comparison')
plt.show()
```

# Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
```

```python
model = Sequential([
    Dense(256, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(16,  activation='relu'),
    Dense(1, activation='linear')  # Output layer with linear activation for regression
])
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mse'])
```

```python
model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=250, batch_size=32, verbose=1)
```

```python
evaluation = model.evaluate(x_test, y_test, verbose=1)
print(f"Test Loss: {evaluation[0]}, Test MAE: {evaluation[1]}")
```
